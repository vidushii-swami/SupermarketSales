### Data Analysis 

We will separate the data up into categories. Visualizations will help understand the type of people who shop, the ways they pay and what they pay for. 


```python
CType = supermarket["Customer type"].value_counts()
CType
```

```python
Gender = supermarket["Gender"].value_counts()
Gender
```

```python
Payment = supermarket["Payment"].value_counts()
Payment 
```

```python
City = supermarket["City"].value_counts()
City
```

```python
ProductLine = supermarket["Product line"].value_counts()
ProductLine
```

```python
Branch = supermarket["Branch"].value_counts()
Branch
```

```python
gross_income = supermarket.groupby('Product line').agg({'gross income': 'sum','Total': 'mean'})
gross_income = gross_income.rename(columns={'gross income':'Total income','Total':'Avg invoice income'})
gross_income
```

```python
# Set up the figure and the subplots
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(14, 10))

# Plot and annotate the first subplot for Product Line
sns.barplot(x=ProductLine.index, y=ProductLine.values, ax=axs[0, 0], palette="viridis")
axs[0, 0].set_title('Product Line Distribution')
axs[0, 0].set_xlabel('Product Line')
axs[0, 0].set_ylabel('Count')
axs[0, 0].tick_params(axis='x', rotation=45)
for i in axs[0, 0].containers:
    axs[0, 0].bar_label(i)

# Plot and annotate the second subplot for City
sns.barplot(x=City.index, y=City.values, ax=axs[0, 1], palette="inferno")
axs[0, 1].set_title('City Distribution')
axs[0, 1].set_xlabel('City')
axs[0, 1].set_ylabel('Count')
axs[0, 1].tick_params(axis='x', rotation=45)
for i in axs[0, 1].containers:
    axs[0, 1].bar_label(i)

# Plot and annotate the third subplot for Payment
sns.barplot(x=Payment.index, y=Payment.values, ax=axs[1, 0], palette="magma")
axs[1, 0].set_title('Payment Method Distribution')
axs[1, 0].set_xlabel('Payment Method')
axs[1, 0].set_ylabel('Count')
axs[1, 0].tick_params(axis='x', rotation=45)
for i in axs[1, 0].containers:
    axs[1, 0].bar_label(i)

# Plot and annotate the fourth subplot for Gender
sns.barplot(x=Gender.index, y=Gender.values, ax=axs[1, 1], palette="coolwarm")
axs[1, 1].set_title('Gender Distribution')
axs[1, 1].set_xlabel('Gender')
axs[1, 1].set_ylabel('Count')
for i in axs[1, 1].containers:
    axs[1, 1].bar_label(i)

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

bars = ax.bar(gross_income.index, gross_income['Total income'], color='lightgreen', label='Total Income')
ax.set_xlabel('Product Line')
ax.set_ylabel('Total Income')
ax.set_title('Total Income by Product Line')
plt.xticks(rotation=45, ha='right')


for bar in bars:
    height = bar.get_height()
    ax.annotate(f'{height:.2f}',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3),  
                textcoords='offset points',
                ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(Payment, labels=Payment.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)

# Add a title
plt.title('Payment Method Distribution')

# Display the chart
plt.show()
```

### Time Series Analysis 

```python
supermarket['Month']=supermarket['Date'].dt.month  
supermarket['Day']=supermarket['Date'].dt.day_name()
```

```python
day_sm = supermarket.groupby('Day')['Total'].sum().to_frame()
day_sm = day_sm.reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']) 
day_sm.head(10)
```

```python
# Set up the figure and the bar plot
fig, ax1 = plt.subplots(figsize=(10, 6))

# List of days of the week
daywise_day = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Create the bar plot
ax1 = sns.barplot(x=daywise_day, y=day_sm['Total'], data=day_sm, palette='magma')

# Set the title and labels
ax1.set_title('Weekly Sales of Supermarket', size=16)
ax1.set_ylabel('Total Sales', fontsize=14)
ax1.set_xticklabels(daywise_day, fontsize=14)

# Add the counts on top of each bar
for container in ax1.containers:
    ax1.bar_label(container)

# Show the plot
plt.show()
```
